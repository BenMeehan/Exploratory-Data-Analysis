# **KS - Test**

Used to check if 2 distributions are same using hypothesis testing.

```python
import numpy
x=numpy.random.normal(0,1,1000)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp

sns.kdeplot(x)
```

```python
sp.stats.kstest(x,"norm")
```

```python
print("since p-value is 69%, we conclude that x is normal")
```
