# **Importing Libraries**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the data : **Haberman Cancer Survival Dataset**

```python
df=pd.read_csv("haberman.csv",names=['Age','Year Of Operation','Auxillary Nodes','Status'])
```

# **Univariate Analysis**

**Probability Distribution Function and Cumulative Distribution Function**

```python
sns.set_style("whitegrid")

# Plotting the CDF and PDF for the features

count,bins=np.histogram(df['Age'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.title("Age")
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

plt.figure()
count,bins=np.histogram(df['Year Of Operation'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.title("Year Of Operation")
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

plt.figure()
count,bins=np.histogram(df['Auxillary Nodes'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.title("Auxillary Nodes")
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

plt.show()
```

```python
sur=df[df['Status']==1]
```

```python
count,bins=np.histogram(sur['Age'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.title("Age")
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)

plt.figure()
count,bins=np.histogram(df['Auxillary Nodes'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.title("Auxillary Nodes")
plt.plot(bins[1:],pdf)
plt.plot(bins[1:],cdf)
```
